```python
dfc = df.copy()
```

```python
# --- Standardize strings ---
for col in ["property_type", "specific_type", "county", "city", "area_name", "title"]:
    dfc[col] = (dfc[col]
                .astype(str)
                .str.strip()
                .str.replace(r"\s+", " ", regex=True)
                .str.title())
```

```python
# --- Deduplicating ---
# (1) by listing_id (should be unique)
dfc = dfc.drop_duplicates(subset=["listing_id"])
# (2) safety: also drop dupes by url if any
dfc = dfc.drop_duplicates(subset=["url"])
```

```python
# --- Numeric coercion & flags ---
# bedrooms, area, price
dfc["bedrooms"]  = pd.to_numeric(dfc["bedrooms"], errors="coerce")
dfc["area_sqft"] = pd.to_numeric(dfc["area_sqft"], errors="coerce")
dfc["price_ksh"] = pd.to_numeric(dfc["price_ksh"], errors="coerce")
```

```python
# Flag missing price from site ("Price not communicated")
dfc["price_disclosed"] = ~dfc["price_ksh"].isna()
```

```python
# --- Area conversions & derived metrics ---
SQFT_TO_SQM = 0.092903
dfc["area_sqm"] = dfc["area_sqft"] * SQFT_TO_SQM
```

```python
# price per sqm only where both are present and positive
dfc["price_per_sqm"] = np.where(
    (dfc["price_ksh"] > 0) & (dfc["area_sqm"] > 0),
    dfc["price_ksh"] / dfc["area_sqm"],
    np.nan
)
```

```python
# for readability in charts/tables
dfc["price_m"] = dfc["price_ksh"] / 1_000_000
```

```python
# --- Property category (broad buckets) ---
def map_category(row):
    pt, st = (row.get("property_type",""), row.get("specific_type",""))
    text = f"{pt} {st}".lower()
    if "land" in text or "plot" in text:
        return "Land"
    if "apartment" in text or "flat" in text:
        return "Apartment/Flat"
    if "house" in text or "villa" in text or "maisonette" in text or "townhouse" in text or "bungalow" in text:
        return "House/Townhouse"
    if "commercial" in text or "office" in text or "shop" in text or "warehouse" in text or "retail" in text:
        return "Commercial"
    return "Other"
```

```python
#create new column called property_category and applies the map_category function we did above
dfc["property_category"] = dfc.apply(map_category, axis=1)
```

```python
# ---tidy location (common harmonization) ---
county_map = {
    "Kajiado County": "Kajiado",
    "Kiambu County": "Kiambu",
    "Kilifi County": "Kilifi",
    "Kwale County": "Kwale",
    "Machakos County": "Machakos",
    "Muranga County": "Muranga",
    "Nakuru County": "Nakuru",
    "Nyandarua County": "Nyandarua"
}
dfc["county"] = dfc["county"].replace(county_map)
```

```python
#dropping raw_json for a lean dataset.
dfc = dfc.drop(columns = ["raw_json"])
```

```python
dfc.to_csv("properties_clean.csv", index=False)
print("Saved → properties_clean.csv")
```

```python
df_price = dfc[dfc["price_disclosed"]].copy()
```

```python
# Winsorize/filter for charts only (not for the main dataset)
p_low, p_high = df_price["price_ksh"].quantile([0.01, 0.99])
df_plot = df_price[(df_price["price_ksh"] >= p_low) & (df_price["price_ksh"] <= p_high)].copy()

# Same for price_per_sqm where available
pps_low, pps_high = df_price["price_per_sqm"].quantile([0.01, 0.99])
df_plot_pps = df_price[(df_price["price_per_sqm"] >= pps_low) & (df_price["price_per_sqm"] <= pps_high)].copy()
```

```python
# Supply by county & category
supply_by_county = (dfc
    .groupby(["county","property_category"])
    .agg(listings=("listing_id","count"),
         pct_price_disclosed=("price_disclosed", lambda x: np.mean(x)*100))
    .reset_index()
    .sort_values(["listings"], ascending=False))

# Price stats by county (disclosed only)
price_by_county = (df_price
    .groupby("county")
    .agg(median_price_ksh=("price_ksh","median"),
         mean_price_ksh=("price_ksh","mean"),
         listings=("listing_id","count"))
    .reset_index()
    .sort_values("median_price_ksh", ascending=False))

# Price per sqm by county (disclosed + area available)
pps_by_county = (df_price
    .dropna(subset=["price_per_sqm"])
    .groupby("county")
    .agg(median_pps=("price_per_sqm","median"),
         mean_pps=("price_per_sqm","mean"),
         listings=("listing_id","count"))
    .reset_index()
    .sort_values("median_pps", ascending=False))

# Beds vs price (residential only)
res_mask = dfc["property_category"].isin(["Apartment/Flat","House/Townhouse"])
beds_price = (df_price[res_mask]
    .groupby("bedrooms")
    .agg(median_price_ksh=("price_ksh","median"),
         listings=("listing_id","count"))
    .reset_index()
    .sort_values("bedrooms"))

# Top areas by median price (helpful for buyer shortlists)
top_areas = (df_price
    .groupby(["county","city","area_name"])
    .agg(median_price_ksh=("price_ksh","median"),
         listings=("listing_id","count"))
    .reset_index()
    .query("listings >= 10")  # ensure enough data
    .sort_values("median_price_ksh", ascending=False)
    .head(30))

# Export aggregates for BI
supply_by_county.to_csv("supply_by_county.csv", index=False)
price_by_county.to_csv("price_by_county.csv", index=False)
pps_by_county.to_csv("pps_by_county.csv", index=False)
beds_price.to_csv("beds_price.csv", index=False)
top_areas.to_csv("top_areas.csv", index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
# Price distribution (filtered for visibility)
plt.figure()
df_plot["price_m"].plot(kind="hist", bins=50)
plt.xlabel("Price (Million KSh)")
plt.title("Listing Price Distribution (winsorized 1–99%)")
plt.show()
```

```python
# Median price per county (top 15)
plt.figure()
(price_by_county
 .nlargest(15, "median_price_ksh")
 .set_index("county")["median_price_ksh"]
 .plot(kind="barh"))
plt.xlabel("Median Price (KSh)")
plt.title("Top 15 Counties by Median Listing Price")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Median price per sqm by county (top 15)
plt.figure()
(pps_by_county
 .nlargest(15, "median_pps")
 .set_index("county")["median_pps"]
 .plot(kind="barh"))
plt.xlabel("Median Price per sqm (KSh)")
plt.title("Top 15 Counties by Price per sqm")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Bedrooms vs median price (residential)
plt.figure()
plt.plot(beds_price["bedrooms"], beds_price["median_price_ksh"], marker="o")
plt.xlabel("Bedrooms")
plt.ylabel("Median Price (KSh)")
plt.title("Bedrooms vs Median Price (Residential)")
plt.grid(True, alpha=0.3)
plt.show()
```
